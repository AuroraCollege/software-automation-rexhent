```python
# Import dependancies 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
dataset = pd.read_csv('../datasets/taxstats2015individual28countaveragemedianbypostcode.csv')
dataset.head()
```

```python

def y(x):
    return -6*x + 40000

def kelly_y(x):
    ''' This simple function takes a postcode and returns the average net tax. 
    It does a poor job!'''
    return -10*x + 80000

plt.scatter(dataset['Postcode'], dataset['Average net tax'])
plt.xlabel('Postcode')
plt.ylabel('Average net tax')
kelly_predicted_y = kelly_y(dataset['Postcode']) #Here we use the function to predict y values based on x values
predicted_y = y(dataset['Postcode'])
plt.plot(dataset['Postcode'], kelly_predicted_y, color='red')
plt.plot(dataset['Postcode'], predicted_y, color='purple')

kelly_l1_loss = (dataset['Average net tax'] - kelly_predicted_y).abs().sum()
print('Mr Kelly L1 Loss:', kelly_l1_loss)
kelly_l2_loss = ((dataset['Average net tax'] - kelly_predicted_y) ** 2).abs().sum()
print('Mr Kelly L2 Loss:', kelly_l2_loss)
mae = (dataset['Average net tax'] - kelly_predicted_y).abs().mean()
print("Mr Kelly Mean Absolute Error:", mae)
mse = ((dataset['Average net tax'] - kelly_predicted_y) ** 2).mean()
print("Mr Kelly Mean Squared Error:", mse)

l1_loss = (dataset['Average net tax'] - predicted_y).abs().sum()
print('My L1 Loss:', l1_loss)
l2_loss = ((dataset['Average net tax'] - predicted_y) ** 2).abs().sum()
print('My L2 Loss:', l2_loss)
mae = (dataset['Average net tax'] - predicted_y).abs().mean()
print("My Mean Absolute Error:", mae)
mse = ((dataset['Average net tax'] - predicted_y) ** 2).mean()
print("My Mean Squared Error:", mse)


plt.show()
```
